```python
df1 = df1.rename(columns=lambda x: str(x)[:4])
year=set(df1.columns)
year_list=[]
for i in year:
    year_list.append(i)
year_list=sorted(year_list)
```

```python
mean_list=[]
for i in year_list:
    mean_list.append(df1[i].T.mean())
#저거 연도별 ㅕㅇ균 내고싶음 
```

```python
mean=pd.DataFrame(mean_list,index=year_list)
mean.T
```

```python
df
```

```python
df2=pd.concat([df,mean.T])
df2=df2.T
df2
```

```python
mean = df2.mean()
std = df2.std()
normalized_data = (df2 - mean) / std
normalized_data
# Min-Max Scaling
df_norm = (df2 - df2.min()) / (df2.max() - df2.min())

# 결과 출력
```

```python
df_norm
```

```python
correlations = df_norm.corr()['소비자심리지수']
print(correlations)
```

```python
# 각 열과 소비자 심리 지수 간의 가중 상관관계 계산
weighted_corr = df_norm.corrwith(df_norm['소비자심리지수'])

# 결과 출력
print(weighted_corr)
```
